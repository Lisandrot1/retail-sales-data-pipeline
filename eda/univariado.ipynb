# Análisis Univariado

En esta sección vamos a realizar un análisis univariado de las variables para comprender la distribución y características individuales de cada cliente.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../data/processed/SuperStore_limpio.csv')
```

# Análisis de Modos de Envio

## Descripción
Analizamos la variable `Ship_Mode` que representa el método de envío seleccionado por los clientes para la entrega de sus pedidos. Este análisis nos permitirá entender las preferencias de envío y la distribución de los diferentes modos de entrega.

```python
df_mode = (df['Ship_Mode'].value_counts(normalize=True) * 100).round(2)


fig_mode , ax_mode = plt.subplots(figsize =(7,5))
ax_mode.bar(df_mode.index, df_mode.values, color="skyblue")
ax_mode.grid(axis='y')
ax_mode.set_title('Distribucion de Envios con los Clientes')
df_mode
```

### Observaciones Principales

**Segmento Dominante:** `Standard Class`
- Es la categoría con mayor representación en la base de clientes
- Constituye el grupo más numeroso y probablemente el más rentable
- Representa la columna vertebral del negocio en términos de volumen de clientes

# Análisis de Segmentos de Cliente

## Descripción
Analizamos la variable `Segment` que clasifica a los clientes en diferentes categorías según su perfil de consumo o características demográficas. Este análisis nos permitirá entender la composición de nuestra base de clientes por tipo de segmento.

```python
df_segment = (df['Segment'].value_counts(normalize=True) * 100).round(2)

fig_segment, ax_segment = plt.subplots(figsize=(7,5))
ax_segment.pie(df_segment.values, labels=df_segment.index.to_list(),autopct='%1.2f%%',startangle=90)
ax_segment.set_title('Distribucion Segmento Clientes')

```

**Segmento Dominante:** `Consumer`
- Representa el **51.94%** de la base total de clientes
- Más de la mitad de todos los clientes pertenecen a este segmento
- Constituye el mercado principal y más significativo del negocio

```python
df_mas_city = df['City'].value_counts().head(5)
df_menos_city = df['City'].value_counts().tail(5)

fig, ax = plt.subplots(figsize=(15,7))
ax.bar(df_mas_city.index, df_mas_city.values, color='skyblue')
ax.grid(axis=('y'))
ax.set_title('Ciudades con más clientes de United States')
df_menos_city
```

## Análisis de Concentración Geográfica

### Observaciones Principales

**Ciudad Líder:** `New York City`
- Destaca como la ciudad con mayor concentración de clientes
- Representa el principal centro de operaciones y mercado
- Supera significativamente a las demás ciudades en volumen de clientes

## Descripción
Analizamos la variable `Category` que clasifica los productos en diferentes categorías principales. Este análisis nos permitirá entender la distribución de ventas por tipo de producto y identificar las categorías más populares.

```python
df_category = (df['Category'].value_counts(normalize=True)* 100).round(2)
fig_category, ax_category = plt.subplots(figsize=(7,5))

ax_category.pie(df_category.values, labels=df_category.index.to_list(),autopct='%1.2f%%',startangle=90)
ax_category.set_title('Distribucion de Categorias')
df_category
```

### Observaciones Principales

**Categoría Dominante:** `Office Supplies`
- Representa el **60.30%** de todas las categorías
- Más de la mitad del portafolio pertenece a suministros de oficina
- Constituye la línea de productos más significativa del negocio

```python
df_top_sales = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)

fig_sales, ax_sales = plt.subplots(figsize=(10,5))
ax_sales.barh(df_top_sales.index, df_top_sales.values)
ax_sales.set_title('Productos Con mas Ingresos')
```

### Observaciones Principales

**Producto Más Rentable:** `High Speed Automatic Electric Letter Opener`
- Genera los mayores ingresos entre todos los productos
- Posicionado como el artículo de mayor valor en el portafolio
- Representa el producto estrella en términos de rentabilidad
